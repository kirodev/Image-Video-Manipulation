*   both the images should be placed at the center of square image of 1024x1024 each
*   stack them horizontally, so that you get an output of H x W (1024 x 2048) image
*   output should be a png file
*    transparency of the original png should be maintained.

##Function implementing all the points in Question1 and returning the Horizontally stacked image dimensions as (1024x2048)

```python
def image_stacking(PATH_TO_IMG1, PATH_TO_IMG2, PATH_TO_BKGRND_IMG):
  """
  PATH_TO_IMG1: Path to image 1
  PATH_TO_IMG2: Path to image 2
  PATH_TO_BKGRND_IMG: Path to background image
  return: stacked image dimension

  """
  image1 = cv2.imread(PATH_TO_IMG1, cv2.IMREAD_UNCHANGED) #reading images and checking their original shapes
  print(f'original image1 shape: {image1.shape}')
  
  image2 = cv2.imread(PATH_TO_IMG2, cv2.IMREAD_UNCHANGED)
  print(f'original image2 shape: {image2.shape}')
  
  background_image = cv2.imread(PATH_TO_BKGRND_IMG, cv2.IMREAD_UNCHANGED)
  print(f'original background shape: {background_image.shape}')

  
  image1_rgba = cv2.cvtColor(image1, cv2.COLOR_BGR2RGBA) #converting the images into 4 channel(BGRA/RGBA) images and comparing shapes
  print(f'image1_rgba shape: {image1_rgba.shape}')
  image2_rgba = cv2.cvtColor(image2, cv2.COLOR_BGR2RGBA)
  print(f'image2_rgba.shape: {image2_rgba.shape}')
  background_image_rgba = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGBA)
  print(f'background_image_rgba.shape: {background_image_rgba.shape}')

  
  background_image_rgba = cv2.resize(background_image_rgba, (1024,1024)) #resizing the background to required dimension
  
  background_image_rgba_1 = background_image_rgba.copy()
  background_image_rgba_2 = background_image_rgba.copy()

  #co-ordinates of image1, to be placed at the center of background squared image
  x1_offset = (background_image_rgba_1.shape[1]//2) - (image1_rgba.shape[1]//2)
  x1_end = (background_image_rgba_1.shape[1]//2) + (image1_rgba.shape[1]//2)
  y1_offset = (background_image_rgba_1.shape[0]//2) - (image1_rgba.shape[0]//2)
  y1_end = (background_image_rgba_1.shape[0]//2) + (image1_rgba.shape[0]//2)
  

  #co-ordinates of image2, to be placed at the center of background squared image
  x2_offset = (background_image_rgba_2.shape[1]//2) - (image2_rgba.shape[1]//2)
  x2_end = (background_image_rgba_2.shape[1]//2) + (image2_rgba.shape[1]//2)
  y2_offset = (background_image_rgba_2.shape[0]//2) - (image2_rgba.shape[0]//2)
  y2_end = (background_image_rgba_2.shape[0]//2) + (image2_rgba.shape[0]//2)

  #masking the backgroung image with image1 and image2 at desired positions
  background_image_rgba_1[y1_offset:y1_end, x1_offset:x1_end] = image1_rgba
  background_image_rgba_2[y2_offset:y2_end, x2_offset:x2_end] = image2_rgba
  
  #horizontally stacking the image
  horizontal_stacked_image = cv2.hconcat([background_image_rgba_1,background_image_rgba_2])
  
  cv2.imwrite('final_image_Q1.png',cv2.cvtColor(horizontal_stacked_image, cv2.COLOR_BGR2RGBA))
  
  return horizontal_stacked_image.shape

  


```

#paths for the required images

```python
BKGRND_IMG = '/content/drive/MyDrive/Drive_Analytics_Assignment/black_square_image.png'
IMG1 = '/content/drive/MyDrive/Drive_Analytics_Assignment/CheeseCakeFactory.png'
IMG2 = '/content/drive/MyDrive/Drive_Analytics_Assignment/Petronas.jpeg'
```

#calling the function

```python
print(image_stacking(IMG1, IMG2, BKGRND_IMG))
```
