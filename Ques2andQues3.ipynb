##Function to crop each frame with bbox of personID:1 and create video with the cropped images.

```python
def video_inference(video_file):
  
  cap = cv2.VideoCapture(video_file)
  out_path = os.path.join('/content/drive/MyDrive/Drive_Analytics_Assignment/', 'cropped_person10.webM')
  size = (1280,720)
  fps = 54
  fourcc = cv2.VideoWriter_fourcc('V','P','8','0')
  out = cv2.VideoWriter(out_path, fourcc, fps, size)
  count = 0
 
  while cap.isOpened():
    ret, image_np = cap.read()
    if not ret:
      break
    #shape of each frame
    im_height = image_np.shape[0]
    im_width = image_np.shape[1]
    
    ##bbox of person:1 
    xmin = df1.iloc[count]['xmin']
    ymin = df1.iloc[count]['ymin']
    xmax = df1.iloc[count]['xmax']
    ymax = df1.iloc[count]['ymax']
    
    if xmin is not None:
      left, right, top, bottom = int(xmin), int(xmax), int(ymin), int(ymax)
      cropped_image = image_np[top:bottom, left:right]
      image = cv2.resize(cropped_image, (640,720))
      mask = image_np.copy()
      mask[:] = 0 #creating a black image of same size as original frame in order to mask the cropped image
      mask[0:720,  320:960] = image
      out.write(mask)

    count+=1
    if count == 270:
      break

  out.release()

```

```python
video_inference('/content/drive/MyDrive/Drive_Analytics_Assignment/ques2.mp4')
```

#Question 3:
---Create the mask using a bounding box and remove the background using the mask.

##Function to create mask of personID:1 at the exact location in the original video.

```python
def video_inference1(video_file):
  cap = cv2.VideoCapture(video_file)
  out_path = os.path.join('/content/drive/MyDrive/Drive_Analytics_Assignment/', 'cropped_person1_mask1_video.webM')
  size = (1280,720)
  fps = 54
  fourcc = cv2.VideoWriter_fourcc('V','P','8','0')
  out = cv2.VideoWriter(out_path, fourcc, fps, size)
  count = 0
  
  while cap.isOpened():
    ret, image_np = cap.read()
    if not ret:
      break
    im_height = image_np.shape[0]
    im_width = image_np.shape[1]
    

    xmin = df1.iloc[count]['xmin']
    ymin = df1.iloc[count]['ymin']
    xmax = df1.iloc[count]['xmax']
    ymax = df1.iloc[count]['ymax']
    
    if xmin is not None:
      left, right, top, bottom = int(xmin), int(xmax), int(ymin), int(ymax)
      cropped_image = image_np[top:bottom, left:right]
      mask = image_np.copy()
      mask[:] = 0
      mask[top:bottom, left:right] = cropped_image
      mimage = cv2.resize(mask, (1280,720))
      out.write(mimage)

    count+=1
    if count == 270:
      break
      
  out.release()

```

```python
video_inference1('/content/drive/MyDrive/Drive_Analytics_Assignment/ques2.mp4')
```
